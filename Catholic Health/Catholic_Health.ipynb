# Do clusterigng on one-hot encoded data

```python
# Applying PCA function on training
# and testing set of X component
from sklearn.decomposition import PCA

# get the 3 Principal Components
pca = PCA(n_components=3)

# Map data to the principal compomponents
pca_data = pd.DataFrame(columns = ['PC1','PC2','PC3'], data = pca.fit_transform(dummies))

# Explained Variance By the three principal components
explained_variance = pca.explained_variance_ratio_

```

```python
# Elbow method to determine the appropiate number of clusters
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
from yellowbrick.cluster import KElbowVisualizer
# Quick examination of elbow method to find numbers of clusters to make.
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(pca_data)
Elbow_M.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import mplcursors


X = pca_data.values

# Initialize and fit the KMeans model
k = 5  # Number of clusters

agg_clustering = AgglomerativeClustering(n_clusters=k)
labels = agg_clustering.fit_predict(X)

# Predict cluster labels

# Add cluster labels to the DataFrame
pca_data['cluster'] = labels

# Visualize the clusters (optional)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

colors = ['r', 'g', 'b', 'y', 'c', 'm']  # Adjust as needed for more than 3 clusters
for i in range(k):
    bla = pca_data[pca_data['cluster'] == i]

    scatter = ax.scatter(bla['PC1'], bla['PC2'], bla['PC3'], c=colors[i], label=f'Cluster {i}')
    mplcursors.cursor(scatter, hover=True).connect("add", lambda sel: sel.annotation.set_text(bla.iloc[sel.target.index]['Transaction Description']))

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
plt.title('KMeans Clustering')
plt.legend()
plt.show()

```
